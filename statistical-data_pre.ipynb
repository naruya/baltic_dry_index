```python
import pandas as pd
import numpy as np

df = pd.read_csv("/home/nama/Downloads/COAL Historical Data (1).csv")
month = {1:"Jan", 2:"Feb", 3:"Mar", 4:"Apr", 5:"May", 6:"Jun", 7:"Jul", 8:"Aug", 9:"Sep", 10:"Oct", 11:"Nov", 12:"Dec"}
```

```python
import datetime
from datetime import timedelta
from collections import OrderedDict

start_date = datetime.datetime(2016,1,1)
day_count = 960 # 959
days = []
for single_date in (start_date + timedelta(n) for n in range(day_count)):
    days.append(month[single_date.month]+" "+"{:02}".format(single_date.day)+", "+str(single_date.year))

data = OrderedDict(zip(days, [None]*day_count))
data
```

```python
for index, row in df.iterrows():
    data[row[0]] = row[1]
```

```python
data
```

```python
# data['Aug 16, 2014'] = 363.7
# data['Aug 17, 2014'] = 363.7
```

```python
for day in days:
    if not data[day] == None:
        d = data[day]
    else:
        data[day] = d

data
```

```python
import csv

with open('coal_Jan-01-2016_Aug-17-2018.csv', 'w') as f:
    writer = csv.writer(f, lineterminator='\n') # \n
    writer.writerow(["date","price"])
    for d in data.items():
        writer.writerow([d[0],str(d[1])])
```
